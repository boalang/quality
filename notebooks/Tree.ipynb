```python
from ete3 import NCBITaxa, PhyloTree, Tree, faces, AttrFace, TreeStyle, TextFace
ncbi = NCBITaxa()

def layout(node):
    # If node is a leaf, add the nodes name and a its scientific name
    nameFace = faces.AttrFace("sci_name", fsize=16, fgcolor="#009000")

    rank = 100
    if node.is_leaf():
        faces.add_face_to_node(nameFace, node, column=0)
        #node.add_face(TextFace(rank), column=2, position = "branch-right")



tree = ncbi.get_topology([9825, 9823,9925,30538 ], intermediate_nodes=False)

ts = TreeStyle()
ts.show_leaf_name = False
ts.branch_vertical_margin = 10 # 10 pixels between adjacent branches
ts.layout_fn = layout

tree.render("%%inline", w=100, units="mm", tree_style=ts)
```

```python
def generate_tree(tax_list):
    tree = ncbi.get_topology(tax_list, intermediate_nodes=False)


    # custom layout: adds "rank" on top of branches, and sci_name as tip names
    def my_layout(node):
        if getattr(node, "rank", None):
            rank_face = AttrFace("rank", fsize=7, fgcolor="indianred")
            node.add_face(rank_face, column=0, position="branch-top")
        if node.is_leaf():
            sciname_face = AttrFace("sci_name", fsize=9, fgcolor="steelblue")
            node.add_face(sciname_face, column=0, position="branch-right")
    ts = TreeStyle()
    ts.layout_fn = my_layout
    ts.show_leaf_name = False
    # tree.render("tree.png", tree_style=ts)
    tree.render("%%inline", w=200, units="mm", tree_style=ts)

generate_tree(tax_list)    
```
